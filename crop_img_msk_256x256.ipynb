```python
# este codigo une las imagenes una al lado de la otra para hacer pix to pix translation

import os
import sys
from PIL import Image, ImageEnhance
Image.MAX_IMAGE_PIXELS = None

#myfolder1 = 'C:/Users/john/Downloads/PythonCourseUN/software/datasets/imgjpg/'
#myfolder2 = 'C:/Users/john/Downloads/PythonCourseUN/software/datasets/mskjpg/'
#myoutfolder = "C:/Users/john/Downloads/PythonCourseUN/software/datasets/combined/"

myfolder1 = 'C:/Users/john/Downloads/AutoMap/datos/mos/'
myfolder2 = 'C:/Users/john/Downloads/AutoMap/datos/msk/'
myoutfolder = "C:/Users/john/Downloads/AutoMap/datos/dataset600/"

#total_width = 512
#max_height = 256
total_width = 1200 
max_height = 600

lst1 = []
lst2 = []
#formato = '.jpg'
formato = '.png'

new_im = Image.new('RGB', (total_width, max_height), color=(0,0,0))

for count, filename in enumerate(os.listdir(myfolder1)):
    (lst1.append(filename))

for count2, filename2 in enumerate(os.listdir(myfolder2)):
    (lst2.append(filename2))

for i in range(0,len(lst1)):
    im=Image.open(myfolder1+str(lst1[i])).convert("RGB")
    img=Image.open(myfolder2+str(lst2[i])).convert("RGB")
    new_im.paste(im, (0,0))
    new_im.paste(img, (600,0))
    new_im.save(myoutfolder + str(i+1) + formato)
print("se han emparejado " + str(i+1) + ' imagenes de ' +
      str(max_height) +' pixeles en el formato: ' + formato)
```

```python
# Este script realiza dos transformaciones sobre el dataset emparejado para aumentar los datos
# Down to bottom (flip top to bottom) que equivale a una rotación de 90 grados y
# Iluminación a 40% mas, 1.4

import os
import sys
from PIL import Image, ImageEnhance
Image.MAX_IMAGE_PIXELS = None

mypath = "C:/Users/john/Downloads/AutoMap/datos/dataset600/"
imagenes = 1

for f in os.listdir(mypath):
    i = Image.open(os.path.join(mypath, f))
    print(f, i.size, i.mode)
    fname, fext = os.path.splitext(f)
    i.transpose(Image.FLIP_TOP_BOTTOM).save(os.path.join(mypath,'v{}.png'.format(fname)))
    enhancer = ImageEnhance.Brightness(i)
    enhancer.enhance(1.4).save(os.path.join(mypath,'e{}.png'.format(fname)))
        
    imagenes = imagenes + 1

print("El numero de imagenes procesadas fue de: "+ str((imagenes-1)*2))
```

```python
# Codigo para extraer del dataset grande 10% para validacion y 10% de los datos para test

import os
source = "C:/Users/john/Downloads/AutoMap/datos/dataset600"
dest1 = "C:/Users/john/Downloads/AutoMap/datos/dataset600/valid"
dest2 = "C:/Users/john/Downloads/AutoMap/datos/dataset600/test"

files = os.listdir(source)
import shutil
import numpy as np
for f in files:
    if np.random.rand(1) < 0.05:
        shutil.move(source + '/'+ f, dest1 + '/'+ f)

files = os.listdir(source)
for f in files:
    if np.random.rand(1) < 0.05:
        shutil.move(source + '/'+ f, dest2 + '/'+ f)

print("Ha terminado el proceso de obtener mas o menos 10% de los datos para valid y 10% pa test")
```
